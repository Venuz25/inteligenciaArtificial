### Gráfica de dos dimensiones (Para el dataset Iris)

```python
import matplotlib.pyplot as plt

# Creamos un arreglo de colores según la clase
colors = np.array(['red', 'blue', 'yellow'])
# dataset2.target contiene valores 0, 1, 2
point_colors = colors[dataset2.target]

_, ax = plt.subplots()
scatter = ax.scatter(dataset2.data[:, 0], dataset2.data[:, 1], c=point_colors)
ax.set(xlabel=dataset2.feature_names[0], ylabel=dataset2.feature_names[1])

# Para la leyenda manual:
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='red', label='setosa'),
    Patch(facecolor='blue', label='versicolor'),
    Patch(facecolor='yellow', label='virginica')
]
ax.legend(handles=legend_elements, loc='lower right', title='Classes')
```

### Gráfica de dos dimensiones (Para el dataset Diabetes)

```python
# Elegimos dos características para los ejes X e Y
x_idx = 0  # primera característica
y_idx = 2  # tercera característica

# Discretizamos el target en 2 clases
n_classes = 2
target_discrete = np.digitize(dataset3.target, bins=np.percentile(dataset3.target, [50]))

# Nombres de las clases
target_names = ['Bajo', 'Alto']  # o cualquier etiqueta que quieras

# Crear el gráfico
_, ax = plt.subplots()
scatter = ax.scatter(
    dataset3.data[:, x_idx],
    dataset3.data[:, y_idx],
    c=target_discrete,
    cmap='viridis'
)

ax.set(
    xlabel=dataset3.feature_names[x_idx],
    ylabel=dataset3.feature_names[y_idx]
)

# Leyenda igual al ejemplo que mostraste
handles, _ = scatter.legend_elements()
ax.legend(handles, target_names, loc="lower right", title="Clases")

plt.show()
```

### Gráfica de tres dimensiones (Para el dataset Iris)

```python
# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA

# Definimos los colores por clase
colors = np.array(['red', 'blue', 'yellow'])
point_colors = colors[dataset2.target]  # Asigna el color según la clase

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(dataset2.data)
scatter = ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=point_colors,  # Usamos los colores definidos
    s=40,
)

ax.set(
    title="First three PCA dimensions",
    xlabel="1st Eigenvector",
    ylabel="2nd Eigenvector",
    zlabel="3rd Eigenvector",
)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

# Leyenda manual
legend_elements = [
    Patch(facecolor='red', label='setosa'),
    Patch(facecolor='blue', label='versicolor'),
    Patch(facecolor='yellow', label='virginica')
]
ax.legend(handles=legend_elements, loc="upper right", title="Classes")

plt.show()
```

### Gráfica de tres dimensiones (Para el dataset Diabetes)

```python
# Discretizamos el target en 3 clases para tener 3 colores (similar al iris)
n_classes = 3
target_discrete = np.digitize(dataset3.target, bins=np.percentile(dataset3.target, [33, 66]))

# Definimos los colores por clase
colors = np.array(['red', 'blue', 'yellow'])
point_colors = colors[target_discrete]  # Asigna el color según la clase discretizada

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

# Aplicamos PCA a los datos de diabetes
X_reduced = PCA(n_components=3).fit_transform(dataset3.data)
scatter = ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=point_colors,  # Usamos los colores definidos
    s=40,
)

ax.set(
    title="First three PCA dimensions - Diabetes Dataset",
    xlabel="1st Eigenvector",
    ylabel="2nd Eigenvector",
    zlabel="3rd Eigenvector",
)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

# Leyenda manual con las nuevas clases
legend_elements = [
    Patch(facecolor='red', label='Baja Progresión'),
    Patch(facecolor='blue', label='Media Progresión'),
    Patch(facecolor='yellow', label='Alta Progresión')
]
ax.legend(handles=legend_elements, loc="upper right", title="Diabetes Progression")

plt.show()
```
