# Tuplas
Es una colección ordenada e inmutable de elementos.

```python
tupla1 = (1,3,5,7)

print('Acceso a la tupla con iterator')
for i in tupla1:
  print(i)

print('Acceso a la tupla con índices')
for i in range(len(tupla1)):
  print(i,tupla1[i])
```

## Constructor de tuplas: tuple(iterable)

```python
lista1 = [1, 2, 3]
tupla2 = tuple(lista1)
print(tupla2)
```

```python
lista1[0] = 5
print(lista1)
```

## Uso de índices con tuplas

```python
# Tupla completa
print('tupla1:\t', tupla1)

# Toma desde el inicio hasta el índice 3
print('tupla1[:3]:\t', tupla1[:3])

# Toma desde el índice 1 en adelante
print('tupla1[1:]:\t', tupla1[1:])

# Toma desde el indice 0 hasta el 1
print('tupla1[0:1]:\t', tupla1[0:1])

# Toma desde el índice 1 y el penultimo elemento
print('tupla1[1:-1]:\t', tupla1[1:-1])

# Revierte el orden de la tupla
print('tupla1[::-1]:\t', tupla1[::-1])

# Hace una copia de la tupla
print('tupla1[:]:\t', tupla1[:])

# Toma elementos de dos en 2 de inicio a fin
print('tupla1[::2]:\t', tupla1[::2])
```

### Ejercicio 4
Prueben cada uno de los metodos siguientes en una lista cualquiera e impriman sus resultados.

```python
#Creamos una lista
nombres = ["Ana", "Luis", "Carlos", "Sofia", "Maria"]
print("Lista original:", nombres)
print()
```

```python
#Método append
nombres.append("Diego")
print("append =", nombres)
print()
```

```python
#Método clear
copiaN = nombres.copy()
copiaN.clear()
print("clear =", copiaN)
print()
```

```python
#Método copy
nombres_copia = nombres.copy()
print("copy =", nombres_copia)
print()
```

```python
#Método count
nombres.append("Luis")
countN = nombres.count("Luis")
print("count =", countN)
print()
```

```python
#Método extend
nombres2 = ["Fernanda", "Jorge", "Luz"]
nombres.extend(nombres2)
print("extend =", nombres)
print()
```

```python
#Método index
indexN = nombres.index("Sofia")
print("index =", indexN)
print()
```

```python
#Método insert
nombres.insert(2, "Ricardo")
print("insert =", nombres)
print()
```

```python
#Método pop
eliminado = nombres.pop(3)
print("Elemento removido:", eliminado)
print("pop =", nombres)
print()
```

```python
#Método remove
nombres.remove("Luis")
print("remove =", nombres)
print()
```

```python
#Método reverse
nombres.reverse()
print("reverse =", nombres)
print()
```

```python
#Método sort
nombres.sort()
print("sort =", nombres)
print()
```

## Comprensión de listas

```python
frutas = ["manzana", "kiwi", "guanabana", "limón", "níspero", "Pomelo", "Mango"]
lista2 = []

#Agrega a la lista2 todas las frutas de la lista frutas que contengan la leta 'a'
for x in frutas:
  if "a" in x:
    lista2.append(x)

print(lista2)
```

```python
#Itera sobre la lista frutas y si un elemento (x) contienen una 'a' lo agerega a la lista2
lista2 = [x for x in frutas if "a" in x]
print(lista2)
```

```python
#Itera sobre la lista de frutas y si un elemento NO conitiene una 'a' lo agrega a la lista2
lista2 = [x for x in frutas if "a" not in x]
print(lista2)
```

```python
#Itera sobre la lista frutas y si un elemento tiene un largo de menos de 6 caracteres lo agrega a la lista2
lista2 = [x for x in frutas if len(x) < 6]
print(lista2)
```

```python
#Recorre numeros del 0-9 y agrega todos aquellos cuyo modulo2 se igual a 0 (pares)
lista2 = [x for x in range(10) if x % 2 == 0]
print(lista2)
```

```python
#Recorre la tupla t1 y agrega todos aquello elementos cuyo tipo sea entero o flotante.
lista2 = [x for x in t1 if type(x) == int or type(x) == float]
print(lista2)
```
